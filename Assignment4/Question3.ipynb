The SAT-GPA dataset contains information of SAT and GPA scores. 
We have to predict GPA score on the basis of SAT score by using Linear Regression
and Polynomial Regression model. Also evaluate the model by using evaluation metrics.

#### import required packages


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("SAT-GPA.csv")
df.head()
```

```python
df.info()
```

```python
df.corr()
```

```python
df['GPA'].value_counts
```

```python
x=df.drop('GPA', axis=1)
y= df['GPA']
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures()
x = poly.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=123456)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
y_true = y_test
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_true,y_pred)
r2
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
```
