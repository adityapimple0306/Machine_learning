
In Churn.csv dataset you have to predict that customer is exited or not based on 
various factors given in dataset.
Implement a KNN,SVM and Naive Bayes model and calculate Accuracy 
by confusion matrix, Calculate accuracy score, precision score, recall score, f1 score.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings(action='ignore')
```

```python
df = pd.read_csv("Churn.csv")
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
df['Geography'] = LabelEncoder().fit_transform(df['Geography'])
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
```

```python
df.info()
```

```python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
df.corr()
```

```python
df = df.drop(['CreditScore', 'Geography', 'Tenure', 'NumOfProducts', 'HasCrCard', 'EstimatedSalary'], axis=1)
```

```python
df.head()
```

```python
df['Exited'].value_counts()
```

```python
from imblearn.over_sampling import SMOTE
x = df.drop('Exited', axis=1)
y=df['Exited']
smote = SMOTE()
x_sampled,y_sampled = smote.fit_resample(x,y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.7,random_state=123456)
```

```python
from sklearn.neighbors import KNeighborsClassifier

def create_knn_model():
    model = KNeighborsClassifier()
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.svm import SVC

def create_svm_model():
    model=SVC()
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.naive_bayes import GaussianNB

def create_naive_bayes_model():
    model = GaussianNB()
    model.fit(x_train,y_train)
    return model
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,precision_score,recall_score,f1_score

def evaluate_model(model_name, model):
    y_pred = model.predict(x_test)
    y_true = y_test

    cm = confusion_matrix(y_true,y_pred)
    accuracy = accuracy_score(y_true,y_pred)
    precision = precision_score(y_true,y_pred,average='macro')
    recall = recall_score(y_true,y_pred,average='macro')
    f1 = f1_score(y_true,y_pred,average='macro')

    return model_name,cm,accuracy,precision,recall,f1
```

```python
models = pd.DataFrame([
    evaluate_model("Naive Bayes", create_naive_bayes_model()),
    evaluate_model("KNN", create_knn_model()),
    evaluate_model("SVM", create_svm_model())
], columns=['Model Name', 'cm', 'accuracy', 'precision', 'recall', 'f1'])
models
```
