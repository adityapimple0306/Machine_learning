```python
## split the data
x = df_train.drop('Salary',axis=1)
y = df_train['Salary']
```

```python
## handeling the imbalanced data
from imblearn.over_sampling import SMOTE
smote = SMOTE()
new_x ,new_y = smote.fit_resample(x,y)
```

Prepare following models to predict Salary:
1. Logistic Regression
2. Naive Bayes
3. KNN
4. SVM
5. Decision Tree
6. Random Forest
7. Adaboost

## Train the model

#### Logistic Regression

```python

from sklearn.linear_model import LogisticRegression
def create_logistic_regression_model():
    model = LogisticRegression()
    model.fit(new_x, new_y)
    return model
```

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

def create_naive_bayes_model():
    model = GaussianNB()
    model.fit(new_x, new_y)
    return model
    
```

#### KNN


```python

from sklearn.neighbors import KNeighborsClassifier

def create_knn_model():
    model = KNeighborsClassifier()
    model.fit(new_x, new_y)
    return model
```

#### SVM

```python
from sklearn.svm import SVC

def create_svm_model():
    model = SVC()
    model.fit(new_x, new_y)
    return model
```

## Decision Tree



```python
from sklearn.tree import DecisionTreeClassifier

def create_decision_tree_model():
    model = DecisionTreeClassifier(max_depth=100)
    model.fit(new_x,new_y)
    return model
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

def create_random_forest_model():
    model = RandomForestClassifier(n_estimators=100)
    model.fit(new_x, new_y)
    return model
```

#### AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier

def create_adaboost_model():
    model = AdaBoostClassifier()
    model.fit(new_x,new_y)
    return model
```

### Load the testing Data

```python
df_test = pd.read_csv("Salary_Test.csv")
df_test.head()
```

## EDA

```python
df_test.info()
```

```python
## Encode the data
from sklearn.preprocessing import LabelEncoder
df_test['workclass'] = LabelEncoder().fit_transform(df_test['workclass'])
df_test['education'] = LabelEncoder().fit_transform(df_test['education'])
df_test['maritalstatus'] = LabelEncoder().fit_transform(df_test['maritalstatus'])
df_test['occupation'] = LabelEncoder().fit_transform(df_test['occupation'])
df_test['relationship'] = LabelEncoder().fit_transform(df_test['relationship'])
df_test['race'] = LabelEncoder().fit_transform(df_test['race'])
df_test['sex'] = LabelEncoder().fit_transform(df_test['sex'])
df_test['native'] = LabelEncoder().fit_transform(df_test['native'])
df_test['Salary'] = LabelEncoder().fit_transform(df_test['Salary'])
```

```python
## Check for the NA values
df_test.isna().sum()
```

```python
## split the data
x_test = df_test.drop('Salary', axis=1)
y_test = df_test['Salary']
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model):
    
    y_pred = model.predict(new_x)
    y_true = new_y
    
    train_accuracy = accuracy_score(y_true, y_pred)
    
    y_pred = model.predict(x_test)
    y_true = y_test
    
    test_accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred,average='macro')
    recall = recall_score(y_true, y_pred, average='macro')
    f1 = f1_score(y_true, y_pred, average='macro')

    return train_accuracy,test_accuracy, precision, recall , f1
```

```python
models = [
    ("Logistic Regression", create_logistic_regression_model()),
    ("Naive Bayes", create_naive_bayes_model()),
    ("KNN", create_knn_model()),
    ("SVM", create_svm_model()),
    ("Decision Tree", create_decision_tree_model()),
    ("Random Forest", create_random_forest_model()),
    ("Adaboost", create_adaboost_model())
]
models
```

```python
performance_data = []
for model_name, model in models:
    train_accuracy, test_accuracy, precision, recall, f1 = evaluate_model(model)
    performance_data.append([
        model_name,
        train_accuracy,
        test_accuracy,
        precision,
        recall,
        f1
    ])

performance_chart = pd.DataFrame(performance_data, columns=['Model Name','Training Accuracy', 'Testing Accuracy', 'Precision', 'Recall', 'F1'])
```
