Prepare association rules using apriori for the following  data sets 
1. book.csv
2. movies.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings(action="ignore")
```

```python
df1 = pd.read_csv("book.csv")
df1.head()
```

```python
from mlxtend.frequent_patterns import apriori
rules = apriori(df1, use_colnames=True, min_support=0.4)
rules
```

```python
df2 = pd.read_csv("movies.csv")
df2.head()
```

```python
df2 = df2.drop(['V1', 'V2', 'V3', 'V4', 'V5'], axis=1)
df2.head()
```

```python
from mlxtend.frequent_patterns import apriori
rules = apriori(df2, min_support=0.5, use_colnames=True)
rules
```
